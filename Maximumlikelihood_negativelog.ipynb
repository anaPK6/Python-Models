```python
import numpy as np
import plotly.graph_objects as go

# Define the original function and its negative
def f(x):
    return -x**2 + 4*x

def neg_f(x):
    return x**2 - 4*x

# Generate x values
x_values = np.linspace(0, 4, 400)

# Calculate y values for both functions
y_values_f = f(x_values)
y_values_neg_f = neg_f(x_values)

# Create the plotly figure
fig = go.Figure()

# Add the original function plot
fig.add_trace(go.Scatter(x=x_values, y=y_values_f, mode='lines', name='f(x) = -x^2 + 4x', line=dict(color='blue')))

# Add the negative function plot
fig.add_trace(go.Scatter(x=x_values, y=y_values_neg_f, mode='lines', name='-f(x) = x^2 - 4x', line=dict(color='red')))

# Add markers for the maximum and minimum points
fig.add_trace(go.Scatter(x=[2], y=[f(2)], mode='markers', name='Max Point (x=2, y=4)', marker=dict(color='blue', size=10)))
fig.add_trace(go.Scatter(x=[2], y=[neg_f(2)], mode='markers', name='Min Point (x=2, y=-4)', marker=dict(color='red', size=10)))

# Update layout
fig.update_layout(title='Original Function f(x) and Negative Function -f(x)',
                  xaxis_title='x',
                  yaxis_title='y',
                  showlegend=True)

# Show the plot
fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Example data points
X = np.array([1, 2, 3, 4, 5])
y = np.array([1.1, 1.9, 3.2, 4.1, 5.1])

# Initial prediction (mean of y)
F0 = np.mean(y)

# Residuals
residuals = y - F0

# Create the plot
fig = go.Figure()

# Scatter plot of the data points
fig.add_trace(go.Scatter(x=X, y=y, mode='markers', name='Data Points'))
fig.add_trace(go.Scatter(x=X, y=[F0]*len(X), mode='lines', name='Initial Prediction (F0)'))

# Plot residuals
for i, res in enumerate(residuals):
    fig.add_trace(go.Scatter(x=[X[i], X[i]], y=[F0, y[i]], mode='lines', name=f'Residual {i+1}'))

# Update layout
fig.update_layout(
    title="Gradient Boosting - Initial Prediction and Residuals",
    xaxis_title="X",
    yaxis_title="Y",
    legend_title="Legend"
)

fig.show()
```

```python
import gensim
from gensim.models import Word2Vec
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Sample text corpus
sentences = [["I", "love", "machine", "learning",],
             ["Word2Vec", "is", "a", "cool", "technique"],
             ["I", "love", "NLP" ,"pytorch","tensorflow"]]

# Train Word2Vec model
model = Word2Vec(sentences, vector_size=10, window=2, min_count=1, workers=4)

# Extract word vectors
words = list(model.wv.key_to_index)
word_vectors = model.wv[words]

# Reduce dimensions with PCA
pca = PCA(n_components=2)
word_vecs_2d = pca.fit_transform(word_vectors)

# Plot word vectors
plt.figure(figsize=(10, 5))
plt.scatter(word_vecs_2d[:, 0], word_vecs_2d[:, 1])

for i, word in enumerate(words):
    plt.annotate(word, xy=(word_vecs_2d[i, 0], word_vecs_2d[i, 1]))

plt.title('Word2Vec Word Embeddings')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```
