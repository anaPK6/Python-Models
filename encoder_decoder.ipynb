```python
import numpy as np
import pandas as pd

# Sample sentences
sentences = [
    "I love machine learning",
    "Machine learning is fun",
    "I love coding"
]

# Create vocabulary
words = set(" ".join(sentences).lower().split())
print("words",words)
vocab = sorted(words)
vocab_size = len(vocab)

# Create a dictionary to map words to indices
word_to_index = {word: i for i, word in enumerate(vocab)}
print("word_to_index",word_to_index)


# Initialize co-occurrence matrix
co_occurrence_matrix = np.zeros((vocab_size, vocab_size))

# Define a context window size
window_size = 1

# Populate the co-occurrence matrix
for sentence in sentences:
    words = sentence.lower().split()
    for i, word in enumerate(words):
        word_idx = word_to_index[word]
        start = max(0, i - window_size)
        end = min(len(words), i + window_size + 1)
        for j in range(start, end):
            if i != j:
                context_word_idx = word_to_index[words[j]]
                co_occurrence_matrix[word_idx, context_word_idx] += 1

# Convert to DataFrame for better visualization
co_occurrence_df = pd.DataFrame(co_occurrence_matrix, index=vocab, columns=vocab)
#print(co_occurrence_df)

# Example: Find similar words to 'love'
love_index = word_to_index['love']
similar_words = co_occurrence_df.iloc[love_index].sort_values(ascending=False)
print("\nWords similar to 'love':")
print(similar_words)
```

```python
def calculate_attention(encoder_states, decoder_state):
    # Step 1: Calculate attention scores (dot product between query and keys)
    attention_scores = np.dot(encoder_states, decoder_state)

    # Step 2: Calculate attention weights (softmax)
    attention_weights = np.exp(attention_scores) / np.sum(np.exp(attention_scores))

    # Step 3: Calculate context vector (weighted sum of encoder hidden states)
    context_vector = np.sum(attention_weights[:, np.newaxis] * encoder_states, axis=0)

    return context_vector

def predict_next_word(context_vector, vocabulary):
    max_similarity = -1
    predicted_word = None

    # Compare context vector with all words in the vocabulary
    for word, embedding in vocabulary.items():
        similarity = np.dot(context_vector, embedding)
        if similarity > max_similarity:
            max_similarity = similarity
            predicted_word = word

    return predicted_word
```
